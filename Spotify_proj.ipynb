# KMeans Clustering

```python
from sklearn.preprocessing import LabelEncoder # for encoding playlist_name and playlist_genre
genre_label=LabelEncoder()
name_label=LabelEncoder()
```

```python
data=df.iloc[:,[7,9]] # playlist genre and name
data
```

```python
a1=len(data["playlist_name"].unique())
a2=len(data["playlist_genre"].unique())
print(f"Number of different playlist names are {a1}")
print(f"Number of different playlist genres are {a2}")
```

```python
data["genre_en"]=genre_label.fit_transform(data["playlist_genre"])
data["name_en"]=name_label.fit_transform(data["playlist_name"])
data
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=6) # creating 6 clusters
```

```python
km.fit(data[["name_en","genre_en"]])
```

```python
km.cluster_centers_  # cluster centers
```

```python
data["cluster_group"]=km.labels_ # creating new column denoting the cluster group for each entry
data
```

```python
import seaborn as sns 
sns.scatterplot(x="name_en",y="genre_en",hue="cluster_group",data=data)
#scatter plot
```
